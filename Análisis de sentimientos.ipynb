```python
#_____IMPORTACIÓN DE BIBLIOTECAS

import re
import operator

#_____DECLARACIÓN INICIAL

#apertura de archivos con los cuales se trabajará
tweets = open("tweets.txt", "r", encoding = "utf-8")
sentimientos = open("sentimientos.txt", "r", encoding = "utf-8")
#declaración de diccionarios
diccionario_tweets = {}
diccionario_sentimientos = {}
palabras_almacenadas = {}
dic_orden_alfabetico = {}
#declaración de patrones de búsqueda par limpieza de los textos a evaluar
patron_CE = re.compile(r'\W+', re.ASCII) #limpieza de caracteres especiales
patron_RT = re.compile(r'^(RT)|(Retweet)', re.I) #limpieza de palbras retweet y RT
patron_USU = re.compile(r'\@\w*', re.ASCII) #limpieza de usuarios
patron_HTTP = re.compile(r'[hH][tT][tT][pP]\\*') #limpieza de HTTP
patron_sin_numeros = re.compile(r'[^A-Za-z]') #limpieza de números
texto_busqueda = "text" 
sustituto = " "
#declaración de variables para posicionar valores dentro del diccionario
count = 0
count_pos = 0
#declaración de variales boleanas para evitar errores mientras se lee el archivo "Tweets.txt"
false = bool
true = bool
null = bool

#________INICIO

#extracción de diccionario con la dupla "sentimiento/valr" contenido en el archivo "sentimientos.txt"
for linea in sentimientos:
    termino, valor = linea.lower().split("\t") 
    diccionario_sentimientos[termino] = int(valor)

#extracción y limpieza de tweets del archivo "tweets.txt"
for linea in tweets: #se recorre cada línea dentro del archivo con un bucle
    texto_busqueda_filtrado = re.findall(texto_busqueda, linea, re.ASCII)  #se busca dentro del texto el campo "Text" en forma de cadena 
    nueva_entrada = eval(linea) #se convierte a diccionario la cadena que contenga el campo "Texto"

    if len(texto_busqueda_filtrado) > 0: #si la cadena tiene una longitud mayor de 0 se entiende que el campo "Texto" existe y se procede a su limpieza
        cadena_de_busqueda = nueva_entrada["text"]  #se obtiene en forma de cadena el valor correspondiente al campo "Texto"
        #se procede a limpiar el texto según los patrones especificados en el apartado de declaración de variables
        cadena_sin_usuario = re.sub(patron_USU, sustituto, cadena_de_busqueda) 
        cadena_sin_caracteres_especiales = re.sub(patron_CE, sustituto, cadena_sin_usuario)
        cadena_sin_RT = re.sub(patron_RT, sustituto, cadena_sin_caracteres_especiales)
        cadena_sin_HTTP = re.sub(patron_HTTP, sustituto, cadena_sin_RT)
        cadena_sin_numeros = re.sub(patron_sin_numeros, sustituto, cadena_sin_HTTP)       
        cadena_final = cadena_sin_numeros.strip() #se limpia de espacios en blanco el inicio y el final de la cadenaa
        count_pos = count_pos + 1 #se declara un contador con la intención de designar un nuevo puesto dentro de diccionario a trabajar en el próximo paso
        
        if len(cadena_final) != 0: #si el largo de la cadena después de las diferentes limpiezas es mayor de 0 se procede a ser almacenado en el nuevo diccionario
            diccionario_tweets[count_pos] = cadena_final #se guarda en un nuevo diccionario la cadena con su nuevo valor o "puesto"
    

#Evaluación de los tweets resultantes almacenados en el diccionario "diccionario_tweets"
for tweets_a_evaluar in diccionario_tweets.values(): #Bucle FOR#1: se inicia un bucle para recorrer cada puesto del diccionario
    cadena = {} #se declara un nuevo diccionario
    a = tweets_a_evaluar.strip() #se eliminan los posibles espacios en blanco al principio y final de la nueva cadena
    cadena_antes_diccionario = a.lower().split(" ") #se divide el contenido de la cadena para ser evaluado
    
#conversión de cada palabra de la cadena en un diccionario con valor inicial de 0
    for u in cadena_antes_diccionario: #Bucle FOR#2: se toma cada valor de la cadena como una clave
        valor_inicio = 0
        cadena[u] = int(valor_inicio) #se declara un diccionario con una dupla "clave/valor" en donde el valor inicial es 0
        #Fin Bucle FOR#2
    cadena_unida = " ".join(cadena) #se une la cadena para imprimir el texto a ser evaluado a modo informativo
    valor_acumulativo_tweet = 0     #inicialización de la variable que acumulará el valor del tweet 
    #impresión en pantalla    
    print("                     ")
    print("Tweet a ser evaluado:")
    print("                     ", cadena_unida)
    print("                     ")
    print("Palabras y valor:")
    print("                    ")
    #Fin Bucle FOR#2
    
    for elemento_en_la_cadena in cadena: #Bucle FOR#3 se compara cada 
                                    #elemento de la cadena con un valor en el diccionario de sentimientos    
        for sentimiento_de_diccionario in diccionario_sentimientos: #Bucle FOR#4 se inicia un tercer bucle en el diccionario     
                                    #en donde se recorre cada sentimiento del diccionario 
            #IF#1:
            if elemento_en_la_cadena == sentimiento_de_diccionario: #comparación del elemento de la cadena con el sentimiento en el diccionario
                                    #si el elemento de la cadena existe en el diccionario se almacena el sentimiento junto a su valor 
                valor_sentimiento = diccionario_sentimientos[sentimiento_de_diccionario]  
                cadena[elemento_en_la_cadena] = valor_sentimiento
                print(sentimiento_de_diccionario, "->" , valor_sentimiento) #impresión en pantalla a modo informativo
                valor_acumulativo_tweet = valor_acumulativo_tweet + valor_sentimiento #suma del valor acumulativo del tweet
                palabras_almacenadas[sentimiento_de_diccionario] = valor_sentimiento #se almacenan en un nuevo diccionario las palabras encontradas
            #Fin IF#
        #Fin Bucle FOR#4
    #IF#2:
    if valor_acumulativo_tweet != 0: #si el valor de la variable "valor_acumulativo_tweet" es diferente de 0 se entiende que el tweet
                                    #tiene palabras con valor del listado sentimiento, por tanto se imprime
        valor_promedio = 0 #inicialización de la variable para calcular el valor promedio
        #impresión en pantalla de los valores encontrados
        print("   ")
        print("Este tweet tiene un valor asociado de:", valor_acumulativo_tweet)    
        print("   ")
        print("Asignación de un valor promedio a las palabras que no fueron puntuadas,", valor_promedio)  
        print("este valor se calcula promediando el valor del tweet entre las cantidades")
        print("de palabras sin valor, por no tener referencia en el listado de sentimientos.")
                        
        #bucle para el conteo de las palabras sin valor en el diccionario para asignarle un valor promedio
        count_palabras_sin_valor = 0
        for elemento_en_la_cadena in cadena: #Bucle FOR#5 para asignar un valor promedio a las palabras sin valor 
                                        #en el listado sentimientos
            valor_del_elemento = cadena[elemento_en_la_cadena] #se crea una nueva variable con el valor de su elemento correspondiente
                        #este se comparara con le intención de contar las palabras sin valor para que no entren 
                        #en el promedio que se realizará de las palabras sin valor
            #IF#3:
            if valor_del_elemento == 0:
                count_palabras_sin_valor = count_palabras_sin_valor + 1 
            #Fin #IF#3
        #cálculo del valor promedio de las palabras sin referencias en el listado de sentimientos
                #OJO: en este partado se realizó un cambio respecto del plantemiento. A manera de sacar un promedio de los puntos
                #se dividirá el valor del tweet entre las palabras que no tienen valor asignado.
        valor_promedio = float
        valor_promedio = valor_acumulativo_tweet / count_palabras_sin_valor #cálculo del valor promedio
        valor_promedio_formato = "{0:.2f}".format(valor_promedio) #limitar el número de decimales del resultado
        #Impresión en pantalla
        print("Su valor es de: ", valor_promedio_formato) 
        print("   ")
        print("Relación de palabras, valor asignado:")
        print("   ")
        #Fin Bucle FOR#5
        #bucle para la asignación del valor promedio a las palabras sin referencias en el listado de sentimientos.
        for elemento_en_la_cadena in cadena: ##Bucle FOR#6 para la asignación del valor promedio a las palabras 
                                #que tengan valor de 0 
            valor_del_elemento = cadena[elemento_en_la_cadena]  #declaraciónd de la variable con el valor del elemento a evaluar
            #IF#4
            if valor_del_elemento == 0: #IFsi el valor del elemento de la cadena es igual a 0 se le asigna el nuevo valor promedio
                cadena[elemento_en_la_cadena] = valor_promedio_formato
                #impresión en pantalla del elemento de la cadena con su valor promedio asignado
                print(elemento_en_la_cadena, "->", valor_promedio_formato)
                print("   ")
            #IF#5
        #Fin Bucle FOR#6
        # Recorrer un diccionario, imprimiendo su clave-valor a manera de resumir el contendido de los valores del tweet
        print("A manera de resumen:")
        print("   ")
        print("Tweet analizado: ", cadena_unida)
        print("   ")
        print("Valor del tweet:", valor_acumulativo_tweet) 
        print("   ")
        print("Listado de las palabras y su valor respectivo:")
        print("   ")
        #Bucle FOR#7
        for k,v in cadena.items():
            print(k, "->", v)
        print("   ")
        print(" ___________ ")
        #Fin Bucle FOR#7
    #ELSE del IF#2   
    else: #excepción de la comparación del valor. Si no hay palabra representada dentro del listado de sentimientos
                #se imprime el siguiente mensaje
        print("                         Este tweet no tiene palabra representadas")
        print("               en el listado de sentimientos, por tanto su valor es de: cero")
        print(" ___________ ")
        print("             ")
    #Fin IF#2:
    #Fin Bucle FOR#3
#impresión en pantalla de las palabras conseguidas en el listado de sentimientos y su respectivo valor.

print("Palabras usadas organizadas alfabéticamente:")
print(" ")

resultado = sorted(palabras_almacenadas.items(), key=operator.itemgetter(0)) #se obtiende del diccionario la clave correspondiente
            #de forma alfabética 
#Bucle FOR#8
for item in resultado: #recorrido de cada elemento de la tupla para organizarlo en un conjunto de "clave/valor"
    nueva_clave = item[0]
    nuevo_valor = item[1]
    dic_orden_alfabetico[nueva_clave] = nuevo_valor
#Fin Bucle FOR#8

#Bucle FOR#9
for x in dic_orden_alfabetico: #impresion en pantalla del nuevo diccionario
    
    print(x, "->", dic_orden_alfabetico[x])    
#Fin Bucle FOR#8
#fin de program
```
